### Function for scraping

```python
# search = gn.search(‘author name’) #search by author name
res_df = pd.DataFrame()
def scraping(df):
    stories = []
    for i in range(len(df)):
        search = gn.search(df.iloc[i].loc['Author_Name'])
        newsitem = search['entries']
        for item in newsitem:
            story = {
                # authorname
                'title': item.title,
                'link': item.link
            }
            stories.append(story)
    return stories
scraping(df)
```
